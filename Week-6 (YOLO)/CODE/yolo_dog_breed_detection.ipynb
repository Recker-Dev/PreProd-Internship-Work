### Import Dependencies

```python
import os
from ultralytics import YOLO
```

### Getting the right path to the yaml file

```python
PATH_TO_YAML = "data/yolo_dog_breed_detector.yaml"
```

### Load YOLO model

```python
#Load a model
model = YOLO('yolo11n.pt')

#Use the model
results = model.train(data=PATH_TO_YAML, epochs=250)
```
