## 1. Sentiment Analysis

### Hugging Face Loads in the Pre-Trained Model for Sentiment Analysis

Model [default]: `distilbert/distilbert-base-uncased-finetuned-sst-2-english`

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
results = classifier(["I hate you", "I love you","Hugging Face makes transformers easy!", "This is not what I expected, it's sub-par at best"])
for result in results:
  print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

## 3. QnA

### Hugging Face Loads in the Pre-Trained Model for Question-Answering

Model: `deepset/roberta-base-squad2` ; we are passing in a particular model instead of the default one

```python
from transformers import pipeline

pipe = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
QA_input = {
    'question': 'How does a piston function in an internal combustion engine?',
    'context': (
        "A piston is a vital component of an internal combustion engine, designed to convert the energy from fuel combustion into mechanical motion. "
        "The piston is a cylindrical part that moves back and forth within the engine's cylinder. This motion occurs in a sequence known as the four-stroke cycle, "
        "which includes the intake, compression, power, and exhaust strokes. During the intake stroke, the piston descends, drawing a mixture of fuel and air into the cylinder. "
        "In the compression stroke, the piston ascends, compressing the fuel-air mixture to prepare it for ignition. The power stroke occurs when a spark ignites the mixture, "
        "resulting in a controlled explosion that drives the piston downward with great force. Finally, during the exhaust stroke, the piston rises again to expel the burned gases "
        "through the exhaust valve. The piston's linear motion is then transferred to the crankshaft via a connecting rod, where it is converted into rotational motion, ultimately "
        "powering the vehicle's drivetrain. Pistons are fundamental to the engine's operation, transforming chemical energy into the mechanical energy that propels vehicles forward."
    )
}

## The model needs a context and a question 
res = pipe(QA_input)
```

```python
## Printing output of the model

print(res["answer"])
```

## 4. Text Summarization

### Hugging Face Loads in the Pre-Trained Model for Text-Summarization

Model: `cnicu/t5-small-booksum` ; we are passing in a particular model instead of the default one

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("cnicu/t5-small-booksum")
model = AutoModelForSeq2SeqLM.from_pretrained("cnicu/t5-small-booksum")
```

```python
inp_text =" The Apollo 11 mission, launched on July 16, 1969, marked a significant milestone in human history as the first successful attempt to land humans on the Moon. Astronauts Neil Armstrong, Buzz Aldrin, and Michael Collins were aboard the spacecraft. On July 20, 1969, Armstrong became the first person to step on the lunar surface, uttering the famous words, (That's one small step for man, one giant leap for mankind.) This achievement showcased the potential of human ingenuity and remains a symbol of exploration and innovation."
```

```python
## Create a token of inp text.
token = tokenizer.encode(inp_text,return_tensors="pt")
# token[0]
```

```python
# Generate a summary using the model
summary_ids = model.generate(token, max_length=57, num_beams=4, early_stopping=True)

# generate ():
# without it, sequence models, like T5, require input both for the encoder (your inp_text) and for the decoder (to generate predictions).
# Handles the need for decoder_input_ids internally.
# You just provide the encoder's input (token) and the function will generate the output text.

# Hyperparams:
# max_length=50: Limits the length of the generated text.
# num_beams=4: Uses beam search to improve the quality of the generated text.
# early_stopping=True: Stops generation when the model produces a complete sentence.
```

```python
# Decode the generated summary into readable text, its otherwise a token list of lists.
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

## skip_special_tokens=True---> eliminates tokens like : <pad><extra_id_0>
```

```python
print("Original Text:")
print(textwrap.fill(inp_text, width=120))
print("-------------------------------------------------------------------------------------")
print("Summarized Text:")
print(textwrap.fill(summary, width=120))
print("-------------------------------------------------------------------------------------")
```
